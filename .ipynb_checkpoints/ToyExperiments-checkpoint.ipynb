```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(8)

class EightGaussiansDataset():
    '''Adapted from https://github.com/rtqichen/ffjord/blob/master/lib/toy_data.py'''
    def _create_data(self,num_points):
        scale = 6.8
        bias = np.pi / 8
        step = np.pi / 4
        centers = [(np.cos(bias + 0*step), np.sin(bias + 0*step)),
                   (np.cos(bias + 1*step), np.sin(bias + 1*step)),
                   (np.cos(bias + 2*step), np.sin(bias + 2*step)),
                   (np.cos(bias + 3*step), np.sin(bias + 3*step)),
                   (np.cos(bias + 4*step), np.sin(bias + 4*step)),
                   (np.cos(bias + 5*step), np.sin(bias + 5*step)),
                   (np.cos(bias + 6*step), np.sin(bias + 6*step)),
                   (np.cos(bias + 7*step), np.sin(bias + 7*step))]
        centers = [(scale * x, scale * y) for x, y in centers]

        dataset = []
        for i in range(num_points):
            point = np.random.randn(2) * 0.5
            idx = np.random.randint(8)
            center = centers[idx]
            point[0] += center[0]
            point[1] += center[1]
            dataset.append(point)
        dataset = np.array(dataset, dtype="float32")
        dataset /= 1.414
        return dataset
    
np.random.seed(8)
Gaussians = EightGaussiansDataset()
generations = Gaussians._create_data(1000)

rho = 0.8
A = np.random.normal(0,2,size=(2,2))
X = np.matmul(generations,A)
simulation = rho * X + np.sqrt(1 - rho**2)*X + np.random.normal(size=X.shape)

fig,ax = plt.subplots(1,3,figsize=(15,5))
ax = ax.ravel()

ax[0].scatter(generations[:,0],generations[:,1],alpha=0.1,color='red',label='Gen.')
ax[0].scatter(simulation[:,0],simulation[:,1],alpha=0.1,color='blue',label='Reco.')
ax[0].legend()

ax[1].scatter(simulation[:,0],simulation[:,1],alpha=0.1,color='blue',label='Reco.')
ax[1].legend()

ax[2].scatter(generations[:,0],generations[:,1],alpha=0.1,color='red',label='Gen.')
ax[2].legend()

plt.show()
```

```python
import os
import json
import argparse
import torch
import random
import numpy as np
from datetime import datetime
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from dataloader.dataloader import CreateLoaders
import pkbar
import torch.optim as optim
from torch.optim import lr_scheduler
import torch.nn as nn
from models.freia_models import create_freai
from dataloader.create_data import return_data
from mmd_loss import MMD

dataset = [gen_features,sim_features]

N_total = len(gen_features)
N_t = int(0.7 * len(gen_features))
N_v = len(gen_features[N_t:])
# Setup random seed
torch.manual_seed(8)
np.random.seed(8)
random.seed(8)
torch.cuda.manual_seed(8)

gen_features = dataset[0]
sim_features = dataset[1]

scaler = MinMaxScaler((-1,1))
scaler.fit(np.concatenate([gen_features,sim_features],axis=0))

gen_features = scaler.transform(gen_features)
sim_features = scaler.transform(sim_features)

train_gen = gen_features[:N_t]
train_sim = sim_features[:N_t]
val_gen = gen_features[N_t:]
val_sim = sim_features[N_t:]

train_dataset = TensorDataset(torch.tensor(train_sim),torch.tensor(train_gen))
val_dataset = TensorDataset(torch.tensor(val_sim),torch.tensor(val_gen))

train_loader = DataLoader(train_dataset,
                    batch_size=1024,
                    shuffle=True)
val_loader =  DataLoader(val_dataset,
                        batch_size=1024,
                        shuffle=False)

history = {'train_loss':[],'val_loss':[],'lr':[]}

print("Training Size: {0}".format(len(train_dataset)))
print("Validation Size: {0}".format(len(val_dataset)))

# Create the model
# This will map gen -> Reco
num_layers = 8
input_shape = gen_features.shape[1]

net = create_freai(input_shape,layers=num_layers)
t_params = sum(p.numel() for p in net.parameters())
print("Network Parameters: ",t_params)
device = torch.device('cuda')
net.to('cuda')

# Optimizer
num_epochs=20
lr = 4e-4

optimizer = optim.Adam(list(filter(lambda p: p.requires_grad, net.parameters())), lr=lr)
num_steps = len(train_loader) * num_epochs
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer=optimizer, T_max=num_steps, last_epoch=-1,eta_min=0)


startEpoch = 0
global_step = 0

run_val = 1.0
print('===========  Optimizer  ==================:')
print('      LR:', lr)
print('      num_epochs:', num_epochs)
print('')


# Define your loss function
cost_fn = nn.L1Loss(reduction='mean')
mse_fn = nn.SmoothL1Loss(reduction='mean')
for epoch in range(startEpoch,num_epochs):

    kbar = pkbar.Kbar(target=len(train_loader), epoch=epoch, num_epochs=num_epochs, width=20, always_stateful=False)

    net.train()
    running_loss = 0.0

    for i, data in enumerate(train_loader):
        # 1 Is Gen, 0 Is Reco
        input_gen  = data[1].to('cuda').float()
        input_reco = data[0].to('cuda').float()
        # reco sampled form dist p, gen sampled from dist q

        optimizer.zero_grad()

        with torch.set_grad_enabled(True):
            gen_from_reco,_ = net.forward(input_reco,rev=False)
            reco_from_gen,_ = net.forward(input_gen,rev=True)


        XX_gen,XY_gen,YY_gen = MMD(gen_from_reco,input_gen,'multiscale')
        XX_reco,XY_reco,YY_reco = MMD(reco_from_gen,input_reco,'multiscale')

        c_p_to_q = cost_fn(gen_from_reco,input_reco) # Cost Function, Gradient is straight line
        c_q_to_p = cost_fn(reco_from_gen,input_gen) # Cost Function, Gradient is straight line

        mse_gen = mse_fn(input_gen,gen_from_reco) # Just keeping track of these
        mse_reco = mse_fn(input_reco,reco_from_gen)

        kl_div = 0.#config['optimizer']['KL_scale']*sum([block.kl_div() for block in net]) / (len(train_loader) * num_layers)

        mmd = (XX_gen + XX_reco).mean() - 2.0*(XY_gen + XY_reco).mean() + (YY_gen + YY_reco).mean()
        cost = (c_q_to_p + c_p_to_q)*1e-4

        loss = 1.2*mmd + kl_div + mse_gen + mse_reco
        #loss = mse_gen + mse_reco

        loss.backward()
        optimizer.step()
        scheduler.step()

        running_loss += loss.item() * input_gen.shape[0]
        kbar.update(i, values=[("loss", loss.item()),("mse_gen",mse_gen.item()),("mse_reco",mse_reco.item()),('MMD',mmd.item()),('cost',cost.item()),('kl_div',kl_div)])
        #kbar.update(i, values=[("loss", loss.item()),("mse_gen",mse_gen.item()),("mse_reco",mse_reco.item()),('kl_div',0.)])
        global_step += 1


    history['train_loss'].append(running_loss / len(train_loader.dataset))
    history['lr'].append(scheduler.get_last_lr()[0])


    ######################
    ## validation phase ##
    ######################
    if bool(run_val):
        net.eval()
        val_loss = 0.0
        val_mmd = 0.0
        val_kl = 0.0
        val_mse_gen = 0.0
        val_mse_reco = 0.0
        val_cost = 0.0
        with torch.no_grad():
            for i, data in enumerate(val_loader):
                input_gen  = data[1].to('cuda').float()
                input_reco = data[0].to('cuda').float()

                gen_from_reco,_ = net.forward(input_reco,rev=False)
                reco_from_gen,_ = net.forward(input_gen,rev=True)

                XX_gen,XY_gen,YY_gen = MMD(gen_from_reco,input_gen,'multiscale')
                XX_reco,XY_reco,YY_reco = MMD(reco_from_gen,input_reco,'multiscale')

                c_p_to_q = cost_fn(gen_from_reco,input_reco) # Cost Function, Gradient is straight line
                c_q_to_p = cost_fn(reco_from_gen,input_gen) # Cost Function, Gradient is straight line


                kl_div = 0.0#config['optimizer']['KL_scale']*sum([block.kl_div() for block in net]) / (len(val_loader) * num_layers)

                mmd = (XX_gen + XX_reco).mean() - 2.0*(XY_gen + XY_reco).mean() + (YY_gen + YY_reco).mean()

                cost = (c_q_to_p + c_p_to_q)*1e-4

                loss = 1.2*mmd + kl_div
                mse_gen = mse_fn(input_gen,gen_from_reco) # Just keeping track of these
                mse_reco = mse_fn(input_reco,reco_from_gen)
                #loss = mse_gen + mse_reco + kl_div

                val_loss += loss
                val_mse_reco += mse_reco
                val_mse_gen += mse_gen
                val_mmd += mmd
                val_kl += kl_div
                val_cost += cost

        val_mmd /= len(val_loader)
        val_cost /= len(val_loader)
        val_loss = val_loss.cpu().numpy() / len(val_loader)
        val_kl /= len(val_loader)
        val_mse_gen /= len(val_loader)
        val_mse_reco /= len(val_loader)

        history['val_loss'].append(val_loss)

        kbar.add(1, values=[("val_loss", val_loss.item()),('val_MMD',val_mmd.item()),('val_cost',val_cost.item()),('val_kl',val_kl)])
        #kbar.add(1, values=[("val_loss", val_loss.item()),('val_mse_reco',val_mse_reco.item()),('val_mse_gen',val_mse_gen.item()),('val_kl',0.0)])

    else:
        kbar.add(1,values=[('val_loss',0.)])
```

```python
net.eval()

pred_gens =[]
true_gens = []

for i,data in enumerate(val_loader):
    reco = data[0].to('cuda').float()
    true_gens.append(data[1].numpy())
    with torch.no_grad():
        out,_ = net.forward(reco,rev=False)
    
    out = out.detach().cpu().numpy()
    
    pred_gens.append(out)

    
pred_gens = scaler.inverse_transform(np.concatenate(pred_gens))
true_gens = scaler.inverse_transform(np.concatenate(true_gens))
```

```python
b = (pred_gens - true_gens) / (true_gens + 1e-50)
```

```python
print(len(np.where(b[:,0] < 0.05)[0])/len(b))
print(len(np.where(b[:,1] < 0.05)[0])/len(b))
```
